```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv('C:/Users/LENOVO/OneDrive/Documents/sales.csv')
data
```

```python
plt.figure(figsize=(10,6))
plt.hist(data['qty'],bins=10,color='skyblue',edgecolor='black')
plt.title('distribution of amount')
plt.xlabel('amount')
plt.ylabel('frequency')
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,6))
axes[0].hist(data['amount'],bins=10,color='skyblue',edgecolor='black')
axes[0].set_title('distribution of amount')
axes[0].set_xlabel('amount')
axes[0].set_ylabel('frequency')
axes[1].hist(data['qty'],bins=10,color='skyblue',edgecolor='black')
axes[1].set_title('distribution of amount')
axes[1].set_xlabel('amount')
axes[1].set_ylabel('frequency')
plt.tight_layout()
                  
plt.show()
```
